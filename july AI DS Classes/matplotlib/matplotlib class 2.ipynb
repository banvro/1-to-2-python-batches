```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from numpy import random
```

```python
sales = np.array([12, 23, 34, 45, 56, 78, 89])
profit = random.randint(10, 15, 7)
```

```python
plt.plot(sales, profit, marker = ".", ms = 16, mec = "red", mfc = "yellow")
```

```python
plt.plot(
    sales, 
    profit, 
    marker = ".", 
    ms = 16, mec = "red", mfc = "yellow", 
    color = "green", linewidth = 2, linestyle = "dashdot")
```

```python
plt.plot(
    sales, 
    profit, 
    marker = ".", 
    ms = 16, mec = "red", mfc = "yellow", 
    color = "green", linewidth = 2, linestyle = "dashdot", label = "Our Profit")
plt.legend()
plt.xlabel("Sales Data")
plt.ylabel("Profit Data")
plt.title("Revenue", color = "red", loc = "left")
```

```python
plt.plot(
    sales, 
    profit, 
    marker = ".", 
    ms = 16, mec = "red", mfc = "yellow", 
    color = "green", linewidth = 2, linestyle = "dashdot", label = "Our Profit")
plt.legend()
plt.xlabel("Sales Data")
plt.ylabel("Profit Data")
plt.title("Revenue", color = "red", loc = "left")

plt.grid()
```

```python
month = np.array(["Jan", "Feb", "March", "April"])
prfot = np.array([10, 14, 12, 15])

co = ["pink", "red", "blue", "yellow"]

plt.bar(month, prfot, color = co, label = "Profit")
plt.plot(month, prfot, marker = "*", ms = 12, color = "green", label = "profit trend line")
plt.legend()
```
