```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
month = np.array([1, 3, 5, 7, 9, 11])
profit = np.array([10, 40, 12, 70, 23, 45])
```

```python
# plot --> draw a line on our datapoints.

plt.plot(month, profit)
```

```python
# markers ==> represnt the datapoint.

plt.plot(month, profit, marker = ".")
```

```python
# marker size ==> ms

plt.plot(month, profit, marker = ".", ms = 20)
```

```python
# change marker color:
    # 1) mec => marker edge  color -----
    # 2) mfc => marker face color

plt.plot(month, profit, marker = ".", ms = 20, mec = "red")
```

```python
plt.plot(month, profit, marker = ".", ms = 20, mec = "red", mfc = "yellow")
```

```python
# markers : *

plt.plot(month, profit, marker = "*", ms = 20, mec = "red", mfc = "yellow")
```

```python
plt.plot(month, profit, marker = "^", ms = 20, mec = "red", mfc = "yellow")
```

```python
# markers => ., *, ^, <, >, 1


plt.plot(month, profit, marker = "*", ms = 20, mec = "red", mfc = "yellow")
```

```python
# properiets on lines
# color = ?

plt.plot(month, profit, marker = "*", ms = 20, mec = "red", mfc = "yellow", color = "green")
```

```python
# linewidth --> chnage thickness of line

plt.plot(month, profit, marker = "*", ms = 20, mec = "red", mfc = "yellow", color = "green", linewidth = 7)
```

```python
# linestyle

plt.plot(month, profit, marker = "*", ms = 20, mec = "red", mfc = "yellow", color = "green",
         linewidth = 3, linestyle = "dotted")
```

```python
plt.plot(month, profit, marker = "*", ms = 20, mec = "red", mfc = "yellow", color = "green",
         linewidth = 3, linestyle = "dashed")
```

```python
plt.plot(month, profit, marker = "*", ms = 20, mec = "red", mfc = "yellow", color = "green",
         linewidth = 3, linestyle = "dashdot")
```

```python
# linestyle : 

# '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'


plt.plot(month, profit, marker = "*", ms = 20, mec = "red", mfc = "yellow", color = "green",
         linewidth = 3, linestyle = "dashdot")
```

```python
# labels

plt.plot(month, profit, marker = "*", ms = 20, mec = "red", mfc = "yellow", color = "green",
         linewidth = 3, linestyle = "dashdot")
plt.xlabel("Months Count")
plt.ylabel("Profit per month")
```

```python
# title

# labels

plt.plot(month, profit, marker = "*", ms = 20, mec = "red", mfc = "yellow", color = "green",
         linewidth = 3, linestyle = "dashdot")
plt.xlabel("Months Count")
plt.ylabel("Profit per month")
plt.title("My Revenue")
```

```python
# grid

plt.plot(month, profit, marker = "*", ms = 20, mec = "red", mfc = "yellow", color = "green",
         linewidth = 3, linestyle = "dashdot")
plt.xlabel("Months Count")
plt.ylabel("Profit per month")
plt.title("My Revenue")
plt.grid()
```

```python
# grid modifcations

plt.plot(month, profit, marker = "*", ms = 20, mec = "red", mfc = "yellow", color = "green",
         linewidth = 3, linestyle = "dashdot")
plt.xlabel("Months Count")
plt.ylabel("Profit per month")
plt.title("My Revenue")
plt.grid(color = "green", linewidth = 0.4, linestyle = "dotted")
```

```python
from numpy import random
```

```python
age = np.arange(11, 101, 10)
profit = random.randint(20, 40, size = (9))
```

```python

age1 = np.arange(12, 102, 10)
profit1 = random.randint(20, 40, size = (9))
```

```python
plt.plot(age, profit, label = "age and profit", marker = ".", ms = 10)
plt.plot(age1, profit1, label = "age1 or prfit 1", marker = "*", ms = 10)
plt.legend(loc = "upper center")

# locations : 'best', 'upper right', 'upper left', 'lower left', 'lower right', 
        #'right', 'center left', 'center right', 'lower center', 'upper center', 'center'
```

```python
month = np.array([23, 33, 50, 67, 78, 100])
profit = np.array([10, 40, 12, 70, 23, 45])
```

```python
plt.plot(month, profit, marker = ".", ms = 12, mec = "red", mfc = "yellow", color = "green")

for i in range(len(month)):
    plt.annotate(xy = (month[i], profit[i]), text = f"({month[i]}, {profit[i]})")
```
