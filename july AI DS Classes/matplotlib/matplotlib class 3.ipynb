```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# bar chart
```

```python
weekdays = np.array(["Monday", "Tuesday", "Wednesday", "Friday", "Saturday", "Sunday"])
```

```python
seals = np.array([20, 40, 25, 42, 23, 43])
```

```python
plt.bar(weekdays, seals)
```

```python
x = ["red", "yellow", "blue", "pink", "lightgreen", "green"]

plt.bar(weekdays, seals, color = x, width = 0.3)
```

```python
x = ["red", "yellow", "blue", "pink", "lightgreen", "green"]

plt.bar(weekdays, seals, color = x, width = 0.8)

for i in range(len(weekdays)):
    plt.annotate(xy = (weekdays[i], seals[i]), text = f"{seals[i]}")
```

```python
# Pie Chart : 
```

```python
profit = np.array([30, 20, 40, 10, ])
```

```python
days = ["Monday", "Tuesday", "Wed", "Friday", "Sunday"]
```

```python
plt.pie(profit, labels = days)
```

```python
# exploid


profit = np.array([30, 20, 40, 10, 40])

days = ["Monday", "Tuesday", "Wed", "Friday", "Sunday"]
exp = [0.2, 0, 0, 0, 0]

plt.pie(profit, labels = days, explode = exp)
```

```python
# exploid


profit = np.array([30, 20, 40, 10, 40])

days = ["Monday", "Tuesday", "Wed", "Friday", "Sunday"]
exp = [0.2, 0, 0, 0.1, 0]

col = ["green", "red", "pink", "lightgreen", "orange"]

plt.pie(profit, labels = days, explode = exp, colors = col)
```
